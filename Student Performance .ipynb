```python
#average of subjects 

print("Average grades by subject:")
print(df[['english.grade','math.grade','sciences.grade','language.grade']].mean())
```

```python
#which nationaly has the higest average grades

avg_by_country = df.groupby('nationality')[['english.grade','math.grade','sciences.grade','language.grade']].mean()

print("Average grade by nationality:")
print(avg_by_country)
```

```python
#sort english grade from higest to lowest

print("sorted by English grade:")
print(avg_by_country.sort_values(by='english.grade', ascending=False).head())
```

```python
#average by gender

avg_by_gender = df.groupby('gender')[['english.grade','math.grade','sciences.grade','language.grade']].mean()

print("Average grades by gender:")
print(avg_by_gender)
```

```python
# add a column for average grade

df['avg_grade'] = df[['english.grade', 'math.grade','sciences.grade']].mean(axis=1)
```

```python
#add a column for average application

df['avg_app_rating'] =df[['portfolio.rating','coverletter.rating','refletter.rating']].mean(axis=1)
```

```python
#preview of updated dataframes 

print(df[['name','nationality','english.grade','math.grade','sciences.grade','avg_app_rating','avg_grade']].head())
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# Plot bar chart for English grades by nationality
plt.figure(figsize=(8,6))
sns.barplot(x=avg_by_country.index, y=avg_by_country['english.grade'])
plt.xticks(rotation=90)
plt.title("Average English Grade by Nationality")
plt.ylabel("Average Grade")
plt.show()
```

```python
#Histogram - Distribution of Average Grades column

df['avg_grade'] = df[['english.grade','math.grade','sciences.grade']].mean(axis=1)

plt.figure(figsize=(8,6))
sns.histplot(df['avg_grade'], bins=20, kde=True)
plt.title("Distribution of Average Grades")
plt.xlabel("Average Grades")
plt.ylabel("Number of Students")
plt.show()
```

```python
#Scatter plot again average by column 

df['avg_app_rating'] = df[['portfolio.rating','coverletter.rating','refletter.rating']].mean(axis=1)

plt.figure(figsize=(8,6))
sns.scatterplot(x=df['avg_grade'], y=df['avg_app_rating'])
plt.title("Grades vs Application Ratings")
plt.xlabel("Average Grade")
plt.ylabel("Average Application Rating")
plt.show()
```

**Report**

-Most students score between 3.2-3.8
The histogram of average grades shows that many of the students cluster in this range.This means the overall performance is fairly consistant

-Japan and USA have strongest average performers
The bar chart of average grades by nationality highlights that students from Japan and USA tend to score higher in English ,Maths and science compared to other groups.

-Application ratings coorelate with gardes 
The scatter plot shows a positive relationship.The students with higher grades also tend to have a stronger applictaions ratings(ie.portfolio,cover letter,and rference letter).This also shows that the acedemic perfomance and applications go hand in hand.
